```python
import base64
import os
from byaldi import RAGMultiModalModel
```

```python
RAG = RAGMultiModalModel.from_pretrained("vidore/colpali-v1.2", verbose=1)
```

```python
RAG.index(
    input_path='./docs/budgetInBrief.pdf',
    index_name='budget',
    store_collection_with_index=True,
    overwrite=True
)
```

```python
query = "Details about Prime Minister's Initiatives?"
#What is the budget allocation for social welfare programs?
#What are the non-tax revenue projections for FY 2024-2025?
#What are the projected customs and duties for FY 2024-2025?"
```

```python
results = RAG.search(query, k=1)
```

```python
results
```

```python
image_bytes = base64.b64decode(results[0].base64)
```

```python
filename = 'result.jpg'
with open(filename, 'wb') as f:
  f.write(image_bytes)
```
